```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = data.drop('Result', axis=1).columns
plots_per_page = 12  # you can set 6, 9, 12, etc.

for i in range(0, len(cols), plots_per_page):
    subset_cols = cols[i:i+plots_per_page]
    plt.figure(figsize=(15, 15))  # Adjust height depending on number of plots
    for j, col in enumerate(subset_cols, 1):
        plt.subplot(4, 3, j)  # 4 rows x 3 columns = 12 plots
        sns.countplot(data=data, x=col)
        plt.xlabel(col, fontsize=10)
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
